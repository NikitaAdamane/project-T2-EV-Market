```python
# Making comparison from the dataset
     

# Plotting the frequencies of the brand in the dataset

plt.figure(figsize = (20 , 5))
sns.barplot(x = 'Brand',y = a,data = data1)
plt.grid(axis = 'y')
plt.title('Brands in the datset')
plt.xlabel('Brand')
plt.ylabel('Frequency')
plt.xticks(rotation = 45)
```

```python
data2 = pd.read_csv("C:/Users/DELL/Desktop/PROJECT 2(EV market research)/StateUTs-wise Operational Charging Stations.csv")
```

```python
data2
```

```python
data2.info()
```

```python
# Extracting data from DataFrame
states = data2['State/UT']
charging_stations = data2['Operational Charging Stations']

plt.figure(figsize=(15,8))

# Plotting the bar chart using sns.barplot
sns.barplot(x=states, y=charging_stations)

# Adding labels and title
plt.xlabel('State/UT')
plt.ylabel('Number of Operational Charging Stations')
plt.title('Operational Charging Stations by State/UT')

# Rotating x-axis labels 
plt.xticks(rotation=90)

# Adding numbers at the top of each bar
for i, value in enumerate(charging_stations):
    plt.text(i, value, str(value), ha='center', va='bottom')

# Display the plot
plt.show()
```

```python
# Extracting data from DataFrame
states = data3['State/UT']
invoice_sales = data3['Number of Invoice/Sales']

plt.figure(figsize=(15,10))

# Plotting the bar chart using sns.barplot
sns.barplot(x=states, y=invoice_sales)

# Adding labels and title
plt.xlabel('State/UT')
plt.ylabel('Number of Invoice/Sales')
plt.title('Invoice/Sales by State/UT')

# Rotating x-axis labels for better readability (optional)
plt.xticks(rotation=90)

# Adding numbers at the top of each bar
for i, value in enumerate(invoice_sales):
    plt.text(i, value, str(value), ha='center', va='bottom')

# Display the plot
plt.show()
```
